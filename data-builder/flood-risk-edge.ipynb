```python
import geopandas as gpd
import os
from pathlib import Path
from dotenv import load_dotenv
import json

# Load environment variables from the .env file
load_dotenv()

# Source directory
source = os.getenv("FLOOD_RISK_PATH")
target_dir = os.getenv("FLOOD_RISK_OUT")

output_path = target_dir + "/parcel-flood-risk.json"
    
if not os.path.exists(output_path):
    
    edges = []
    
    # Read a specific layer from the GDB
    gdf = gpd.read_file(source, layer='ExCon100yr_Parcels_TotalFlooded')

    for idx, row in gdf.iterrows():
        if row['apn']:
            edges.append({
                'source': '1',
                'sourceType': 'FloodScenario',
                'target': row['apn'],
                'targetType': 'LandParcel'
            })


    os.makedirs(target_dir, exist_ok=True)
    
    with open(output_path, 'w', encoding='utf-8') as f:
        json.dump(edges, f, ensure_ascii=False, indent=2)
```
